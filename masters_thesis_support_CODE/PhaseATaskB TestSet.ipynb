```python
import json
import os
```

```python
path = "/backup/bioASQ_test_set/task_B_phaseA"
files = os.listdir(path)

for file in files:
    with open(os.path.join(path, file),"r") as f:
        data = json.load(f)
        for q in data["questions"]: 
            q["documents"] = list(map(lambda x:"http://www.ncbi.nlm.nih.gov/pubmed/"+x.split("/")[-1],q["documents"]))[:10]
    
    with open(os.path.join(path, file),"w") as f:
        json.dump(data,f)
        
        
```
