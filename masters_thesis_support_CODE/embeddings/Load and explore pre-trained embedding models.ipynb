## FASTTEXT model

```python


path = datapath(join(trained_embedding_path,"fasttext","BioWordVec_PubMed_MIMICIII_d200.bin"))

model = FastText.load_fasttext_format(path, full_model=False)
```

```python
MODE = "regex_less_700k_freq"
tk=ph.load_tokenizer(mode=MODE)
```

```python
import numpy as np

fasttext_embedding = {index:model.wv[word]/np.linalg.norm(model.wv[word]) for word,index in tk.word_index.items()}

print(len(fasttext_embedding))
```

```python


f_name = "regex_less_700k_freq_embedding.p"

pickle.dump(fasttext_embedding,open(os.path.join('/','backup','pre-trained_embeddings','word_emb_by_fasttext',f_name),"wb"))
```
